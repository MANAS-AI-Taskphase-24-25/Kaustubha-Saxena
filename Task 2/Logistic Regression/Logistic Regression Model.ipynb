## Feature Scaling

```python
#feature scaling
area=np.array(df['Area'])
df['Area']=(area-area.mean())/(area[area.argmax()]-area[area.argmin()])

Perimeter=np.array(df['Perimeter'])
df['Perimeter']=(Perimeter-Perimeter.mean())/(Perimeter[Perimeter.argmax()]-Perimeter[Perimeter.argmin()])

Major_Axis_Length=np.array(df['Major_Axis_Length'])
df['Major_Axis_Length']=(Major_Axis_Length-Major_Axis_Length.mean())/(Major_Axis_Length[Major_Axis_Length.argmax()]-Major_Axis_Length[Major_Axis_Length.argmin()])

Minor_Axis_Length=np.array(df['Minor_Axis_Length'])
df['Minor_Axis_Length']=(Minor_Axis_Length-Minor_Axis_Length.mean())/(Minor_Axis_Length[Minor_Axis_Length.argmax()]-Minor_Axis_Length[Minor_Axis_Length.argmin()])

Convex_Area=np.array(df['Convex_Area'])
df['Convex_Area']=(Convex_Area-Convex_Area.mean())/(Convex_Area[Convex_Area.argmax()]-Convex_Area[Convex_Area.argmin()])

Equiv_Diameter=np.array(df['Equiv_Diameter'])
df['Equiv_Diameter']=(Equiv_Diameter-Equiv_Diameter.mean())/(Equiv_Diameter[Equiv_Diameter.argmax()]-Equiv_Diameter[Equiv_Diameter.argmin()])
```

```python
y_train=df['Value']
x_train=df[['Area','Perimeter','Major_Axis_Length','Minor_Axis_Length','Convex_Area','Equiv_Diameter','Eccentricity','Solidity','Extent','Roundness','Aspect_Ration','Compactness']].values
#x_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
#y_train = np.array([0, 0, 0, 1, 1, 1])
m,n=x_train.shape
m,n
w_i=np.random.randn(n) * 0.01 
b_i=0
j=[]
```

```python
def accuracy(val,pred_val,m):
    correct=0
    for i in range(m):
        if(val[i]==pred_val[i]):
            correct+=1
    acc=correct/m*100
    return acc
    
```

```python
def sigmoid(z):
    g = 1/(1+np.exp(-z))
    return g
```

```python
def cal_cost(x,y,w,b):
    cost=0
    for i in range(m):
        sig=sigmoid(np.dot(x[i],w)+b)
        cost=cost+(-(y[i]*np.log(sig))-(1-y[i])*np.log(1-sig))   
    cost=cost/m
    return cost
```

```python
def cal_derivative(x,y,w,b):
    m, n = x.shape  
    dj_dw = np.zeros(n)  
    dj_db = 0.0 

    # Compute derivatives
    for i in range(n):
        sum_w = 0
        for j in range(m):
            sum_w += ((sigmoid(np.dot(x[j], w) + b) - y[j]) * x[j, i])
        dj_dw[i] = sum_w / m  

    for j in range(m):
        dj_db += (sigmoid(np.dot(x[j], w) + b) - y[j])
    dj_db /= m  

    return dj_dw, dj_db
```

```python
def gradient_descent(x,y,w_i,b_i,alpha,iterations):
    
    for i in range(iterations):
        dj_dw,dj_db=cal_derivative(x,y,w_i,b_i)
        w_i=w_i-(alpha*dj_dw)
        b_i=b_i-(alpha*dj_db)
        cost=cal_cost(x,y,w_i,b_i)
        j.append(cost)
        print(cost)
    return w_i,b_i
```

```python
alpha=0.4
ite=2000
it=np.arange(ite)
w_final,b_final=gradient_descent(x_train,y_train,w_i,b_i,alpha,ite)
```

## Learning Curve

```python
#Learning Curve
plt.plot(it,j)
```

```python
prediction=np.zeros(m)
classs=[]
sig=0
#if z>0 => 1
#if z<0 => =0
#sig=> Probibility of getting Çerçevelik
for i in range(m):
    pre=np.dot(x_train[i],w_final)+b_final
    sig=sigmoid(pre)
    if(pre>=0):
        prediction[i]=1
        classs.append("Çerçevelik")
        
    else:
        prediction[i]=0
        classs.append("Ürgüp Sivrisi")
    print(prediction[i]," ,  ",sig*100)#PRobibility of getting 1
```

```python
df['Predicted_Class']=classs
df['Predicted_Value']=prediction
```
